```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
# Load the dataset
df = pd.read_csv("ufo-sightings-transformed.csv")
df.head()
```

```python
# Display basic info about the dataset
print(df.info())

# Display summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Drop the 'Locale' column
df_cleaned = df.drop('Locale', axis=1)

# Drop rows with missing values in the 'UFO_shape' column
df_cleaned_final = df_cleaned.dropna(subset=['UFO_shape', 'Region', 'Country_Code', 'Description'])

# Print information about the cleaned dataset
print(df_cleaned_final.info())
```

```python
# Save the cleaned DataFrame to a new CSV file
df_cleaned_final.to_csv('cleaned_ufo_data.csv', index=False)

# Print information about the cleaned dataset
print(df_cleaned_final.info())
```
