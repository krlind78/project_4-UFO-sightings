```python
X = df1.drop(columns='UFO_shape')
y = df1[['UFO_shape']]
```

```python
# X_dummies = pd.get_dummies(X)
```

```python
X['Date_time'] = pd.to_datetime(X['Date_time'])
X['Date_time'] = X['Date_time'].dt.year
X.head()
```

```python
X['date_documented'] = pd.to_datetime(X['date_documented'])
X['date_documented'] = X['date_documented'].dt.year
X.head()
```

```python
len(X['Encounter_Duration'].unique())
```

```python
X = X.drop(columns=['Encounter_Duration','Description'])
X.head()
```

```python
X = X.drop(columns='Country_Code')
```

```python
X.head()
```

```python
X = X[['Date_time','date_documented','Year','Month','Hour','latitude','longitude','length_of_encounter_seconds']]
X.shape
```

```python
y.head()
```

```python
len(y['UFO_shape'].unique())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


# Sample data as a pandas DataFrame



# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to your data and transform it
X_scaled = scaler.fit_transform(X)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
print(y_encoded.shape)
y_reshaped = y_encoded.ravel()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_reshaped, test_size=0.2, random_state=42)

import tensorflow as tf
from tensorflow import keras

num_classes = 29  # Number of classes

model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(num_classes, activation='softmax')  # Use 'softmax' for multi-class classification
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))




```

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Rest of your code for data preprocessing, model building, and splitting

# Define the model
# ...

# Compile the model
# ...

# Train the model and collect training history


# Plot the training loss and validation loss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot the training accuracy and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
# KMeans clustering based on location
```

```python
from sklearn.preprocessing import StandardScaler

kmeansx1 = df1[['latitude','longitude']]
scaler = StandardScaler()
kmeansx1[['latitude', 'longitude']] = scaler.fit_transform(kmeansx1[['latitude', 'longitude']])

from sklearn.cluster import KMeans

# Define the number of clusters (K)
K = 3  # Adjust as needed

# Fit K-Means to the latitude and longitude columns
kmeans = KMeans(n_clusters=K, random_state=0)
kmeansx1['Cluster'] = kmeans.fit_predict(kmeansx1[['latitude', 'longitude']])

```

```python
import matplotlib.pyplot as plt

# Plot the clustered data
plt.scatter(kmeansx1['longitude'], kmeansx1['latitude'], c=kmeansx1['Cluster'], cmap='rainbow', marker='o')
plt.title('K-Means Clustering of UFO Sightings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
kmeansx2 = df1[['latitude','longitude','UFO_shape']]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define a colormap for UFO shapes (you can customize this)
cmap = plt.get_cmap('tab20', len(kmeansx2['UFO_shape'].unique()))

# Create a scatter plot for each unique UFO shape
for shape, color in zip(kmeansx2['UFO_shape'].unique(), cmap.colors):
    subset = kmeansx2[kmeansx2['UFO_shape'] == shape]
    plt.scatter(subset['longitude'], subset['latitude'], label=shape, color=color, s=20)

plt.title('UFO Sightings with Shape Overlay')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add a legend to the plot
plt.legend(title='UFO Shape', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
